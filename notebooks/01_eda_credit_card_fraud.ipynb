#### Charger les librairies

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from pathlib import Path

sns.set(style="whitegrid")
```

#### Charger les données

```python
DATA_PATH = Path("../data/creditcard.csv")

df = pd.read_csv(DATA_PATH)
```

#### Distribution de la cible

```python
df["Class"].value_counts()
```

```python
df["Class"].value_counts(normalize=True) * 100
```

En fraude :

-- la fraude est très rare

-- un modèle naïf qui prédit “pas fraude” tout le temps peut avoir 99 % accuracy

C’est pour ça que l’accuracy est une mauvaise métrique ici

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Class", data=df)
plt.title("Distribution des transactions (0 = Normal, 1 = Fraud)")
plt.show()
```

#### Comparer fraude vs normal

```python
df.groupby("Class")["Amount"].describe()
```

```python
plt.figure(figsize=(8,4))
sns.boxplot(x="Class", y="Amount", data=df)
plt.yscale("log")
plt.title("Transaction Amount by Class (log scale)")
plt.show()
```

#### Distribution temporelle

```python
plt.figure(figsize=(8,4))
sns.histplot(df["Time"], bins=50)
plt.title("Distribution of Time Feature")
plt.show()
```
